```python
df_sales=pd.read_csv(r"C:\Users\admin\Desktop\Ineuron\Dataset\sales_data_final.csv")
```

```python
df_sales.head()
```

```python
def profit_flag(a):
    if a>0:
        return("Positive")
    else:
        return ('Negative')
    
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
#It has created a new table and assign the "define" function into it
```

```python
df_sales.head()
```

```python
df_sales["len_customer_name"]=df_sales['customer_name'].apply(len)
```

```python
def quantity_flag(a):
    if a <10:
        return 'Low'
    elif a>10 and a<20:
        return "Medium"
    else:
        return "High"
```

```python
df_sales["quantity_flag"]= df_sales["quantity"].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales['square_quantity'] =df_sales["quantity"].apply(lambda a : a*2)
#By using lamda funtion use square of quantity
```
